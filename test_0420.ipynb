```python
raw_values = ["10", "20", "abc", "5"]

for raw in raw_values:
    try:
        value = int(raw)

        if value >= 20:
            print(f"{value} → 위험")
        elif value >= 10:
            print(f"{value} → 주의")
        else:
            print(f"{value} → 정상")

    except ValueError:
        print(f"{raw} → 숫자 변환 실패")
```

```python
class Robot:
    def __init__(self):
        self.name = "R1"
        self.battery = 25
    def print_status(self):
        print(f"이름: {self.name}")
        print(f"배터리: {self.battery}")
    def move(self):
        self.battery -= 10
        print(f"이동, 배터리: {self.battery}")

r1 = Robot()
r1.move()
```
