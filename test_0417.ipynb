```python
def add(c, d):
    global a
    a = c + c
    d += d
    return a + d
a = [1, 2] # 100
b = [5, 6] # 200
c = a.copy() # 300
d = b.copy() # 400
result = add(c,d)
print(a)
print(b)
print(result)
print(c)
print(d)
```

```python
speed = [5]
distance = [10]
a = speed.copy()
b = distance.copy()

def test(speed, distance):
    speed += [2]
    distance += [4]
    return speed + distance

ret = test(a, b)
print(speed)
print(distance)
print(ret)
```

```python
# 4단계 미니 과제 내 답안
def speed(speed):
    if speed >= 1.0:
        print("속도: 빠름")
    else:
        print("속도: 정상")

def battery(battery):
    if battery <= 30:
        print("배터리: 충전 필요")
    else:
        print("배터리: 정상")
speed(1)
battery(30)
```

```python
# 4단계 미니 과제 교수님 답안
def checkSpeed(speed):
    ret = ""
    spd_lbl = "Vel"
    if speed > 1.0:
        # print(f"{spd_lbl}: 빠름")
        ret = "빠름"
        return ret
    else:
        # print(f"{spd_lbl}: 정상")
        ret = "정상"
        return ret
def checkBattery(battery):
    if battery > 0.3:
        print("배터리: 정상")
    else:
        print("배터리: 충전 필요")

inp_spd = 2.1
ret = checkSpeed(inp_spd)
print(f"속도: [{ret}]")
checkBattery(0.2)
```
